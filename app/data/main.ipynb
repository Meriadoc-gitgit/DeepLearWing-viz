operations = [

## Documentation de l'Ingénierie des Caractéristiques

Ce document explique les nouvelles colonnes créées au cours du processus d'ingénierie des caractéristiques, ainsi que leurs formules respectives.

### 1. Rapport Portance-Trainée
   - **Nom de la colonne :** `Lift-to-Drag Ratio`
   - **Formule :**  
     $$
     \text{Rapport Portance-Trainée} = \frac{\text{cl}}{\text{cd}}
     $$
   - **Description :**  
     Ce ratio représente l'efficacité aérodynamique, calculée comme le rapport entre le coefficient de portance (`cl`) et le coefficient de trainée (`cd`).


### 2. Angle x Reynolds
   - **Nom de la colonne :** `Angle x Reynolds`
   - **Formule :**  
     $$
     \text{Angle x Reynolds} = \text{angle} \times \text{reynolds}
     $$
   - **Description :**  
     Cette caractéristique est le produit de l'angle d'attaque (`angle`) et du nombre de Reynolds (`reynolds`). Elle combine les effets des deux variables.



### 3. Log Reynolds
   - **Nom de la colonne :** `Log Reynolds`
   - **Formule :**  
     $$
     \text{Log Reynolds} = \ln(\text{reynolds})
     $$
   - **Description :**  
     Cette colonne représente le logarithme naturel du nombre de Reynolds (`reynolds`).


### 4. Reynolds Normalisé
   - **Nom de la colonne :** `Normalized Reynolds`
   - **Formule :**  
     $$
     \text{Reynolds Normalisé} = \frac{\text{reynolds} - \mu(\text{reynolds})}{\sigma(\text{reynolds})}
     $$
   - **Description :**  
     Cette valeur normalise le nombre de Reynolds en soustrayant la moyenne et en divisant par l'écart type.



### 5. Carré de l'Angle
   - **Nom de la colonne :** `Angle^2`
   - **Formule :**  
     $$
     \text{Angle}^2 = (\text{angle})^2
     $$
   - **Description :**  
     Cette caractéristique correspond au carré de l'angle d'attaque (`angle`).



### 6. Valeur Absolue de l'Angle
   - **Nom de la colonne :** `|Angle|`
   - **Formule :**  
     $$
     |\text{Angle}| = |\text{angle}|
     $$
   - **Description :**  
     Cette colonne représente la valeur absolue de l'angle d'attaque (`angle`).

## Explication Pas à Pas de la Fonction `create_stratified_sample`
La fonction `create_stratified_sample` est conçue pour effectuer un **échantillonnage stratifié** sur un DataFrame Pandas contenant des données aérodynamiques. L'objectif est de préserver une représentation équilibrée des différentes sous-catégories du dataset tout en réduisant sa taille.

### Signature
```python
def create_stratified_sample(df, n=10000):
```
- **Arguments** :
  - `df` : Le DataFrame source.
  - `n` : Nombre total de lignes souhaitées après échantillonnage (par défaut, 10 000 lignes).
  
- **Retour** :
  - Un DataFrame échantillonné de manière stratifiée, de taille au plus `n` lignes.


### **1. Stratification**

##### **Création des Bins pour l'Angle et Reynolds**
```python
df['angle_bin'] = pd.qcut(df['angle'], q=5, duplicates='drop')
df['reynolds_bin'] = pd.qcut(df['reynolds'], q=5, duplicates='drop')
```
- **`pd.qcut`** :
  - Divise les valeurs continues en **quantiles** égaux (5 dans cet exemple).
  - Crée des groupes appelés "bins" pour les colonnes :
    - `angle` : Divise les angles en 5 intervalles (par ordre de grandeur).
    - `reynolds` : Divise les valeurs de Reynolds en 5 intervalles.
  - **Paramètre `duplicates='drop'`** : Si les valeurs uniques ne suffisent pas pour former les bins, les bins redondants sont supprimés.

- **Ajout au DataFrame** :
  - `angle_bin` : Colonne catégorique indiquant à quel quantile appartient chaque `angle`.
  - `reynolds_bin` : Colonne catégorique pour les valeurs de `reynolds`.



### **2. Groupement par Catégories**

#### **Définition des Groupes**
```python
sample = df.groupby(['name', 'angle_bin', 'reynolds_bin']).apply(
    lambda x: x.sample(n=max(1, int(n * len(x)/len(df))), random_state=42)
).reset_index(drop=True)
```

- **Groupement** :
  - Le DataFrame est regroupé selon :
    - `name` : Nom du profil aérodynamique.
    - `angle_bin` : Bin de l'angle.
    - `reynolds_bin` : Bin de Reynolds.
  - Chaque groupe représente une sous-catégorie spécifique dans le dataset.

- **Échantillonnage au sein de chaque groupe** :
  - **`apply`** : Applique une fonction sur chaque groupe.
  - **`x.sample`** :
    - Échantillonne des lignes dans chaque groupe.
    - **Nombre d’échantillons** : Proportionnel à la taille relative du groupe dans le dataset initial :
      ```python
      n * len(x)/len(df)
      ```
      - Taille du groupe : `len(x)`
      - Taille totale du DataFrame : `len(df)`
    - Si un groupe est trop petit, au moins 1 ligne est échantillonnée grâce à `max(1, ...)`.
    - La seed `random_state=42` garantit la reproductibilité.

- **Réinitialisation des index** :
  - **`reset_index(drop=True)`** : Supprime les index intermédiaires générés par le groupement.



### **3. Nettoyage**

#### **Suppression des Colonnes Temporaires**
```python
sample = sample.drop(['angle_bin', 'reynolds_bin'], axis=1)
```
- Les colonnes `angle_bin` et `reynolds_bin`, créées uniquement pour la stratification, sont supprimées.



### **4. Ajustement Final de la Taille**

#### **Limitation à `n` Lignes**
```python
return sample.sample(n=min(n, len(sample)), random_state=42)
```
- Si le DataFrame échantillonné contient plus de `n` lignes, il est réduit à `n` lignes supplémentaires par échantillonnage aléatoire.
- Si le DataFrame contient déjà moins de `n` lignes, il est retourné tel quel.



### **Résumé :**

La fonction :
1. Crée des bins pour les colonnes `angle` et `reynolds`.
2. Regroupe les données par `name`, `angle_bin`, et `reynolds_bin`.
3. Échantillonne proportionnellement dans chaque groupe pour préserver la distribution originale.
4. Supprime les colonnes temporaires utilisées pour la stratification.
5. Limite le DataFrame final à `n` lignes au maximum.
